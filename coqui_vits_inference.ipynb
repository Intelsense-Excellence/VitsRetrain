```python
import torch
from TTS.api import TTS
from huggingface_hub import snapshot_download

device = "cuda" if torch.cuda.is_available() else "cpu"

MODEL_DIR = "intelsense/vits-bn-ft-single-nctb-f1"

snapshot_download(MODEL_DIR, local_dir=MODEL_DIR)
tts = TTS(model_path=f"{MODEL_DIR}/best_model.pth", config_path=f"{MODEL_DIR}/config.json").to(device)
```

```python
from IPython.display import Audio
import pybangla
nrml = pybangla.Normalizer()

def inference(text):
    text = nrml.text_normalizer(text)
    return tts.tts(text)

text = "বর্তমানে উপসচিব পদে পদোন্নতিতে প্রশাসন ক্যাডার থেকে ৭৫ শতাংশ ও অন্য ক্যাডার থেকে ২৫ শতাংশ নেওয়া হয়। এ ২৫ শতাংশের ক্ষেত্রে অন্যান্য ক্যাডার থেকে আবেদন আহ্বান করে জনপ্রশাসন মন্ত্রণালয়। সুপিরিয়র সিলেকশন কমিটির (এসএসবি) মাধ্যমে যাচাই-বাছাই করে উপসচিব পদে পদোন্নতি দেওয়া হয়।"
# tts.tts_to_file(text=text, file_path="output.wav")
waveform = inference(text)
Audio(waveform, rate=16000)
```
