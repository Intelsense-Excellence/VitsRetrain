```python
import torch
from TTS.api import TTS
from huggingface_hub import snapshot_download

device = "cuda" if torch.cuda.is_available() else "cpu"

MODEL_DIR = "intelsense/vits-bn-ft-single-nctb-f1"

snapshot_download(MODEL_DIR, local_dir=MODEL_DIR)
tts = TTS(model_path=f"{MODEL_DIR}/best_model.pth", config_path=f"{MODEL_DIR}/config.json").to(device)
```

```python
from IPython.display import Audio

text = "তখন কোম্পানির বিনিয়োগকারী সম্পর্ক বিভাগের প্রধান নিশিত ডেভ বলেন, ‘আমরা আশা করছি বকেয়া পরিশোধের বিষয়টি আরও খারাপ দিকে যাবে না। এখন পর্যন্ত আমরা সেই বিকল্পের দিকে যাচ্ছি না। তবে প্রয়োজন হলে আমরা তা বিবেচনা করব।’"
tts.tts_to_file(text=text, file_path="output.wav")
Audio('output.wav')
```
